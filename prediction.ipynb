```python
import yfinance as yf
import yfinance.utils as yfu

# Function to get ticker symbol from company name
def get_ticker_from_name(company_name):
    try:
        # This fetches ticker using yfinance utils for better matching
        tickers = yfu.search(company_name)
        if tickers:
            return tickers[0]['symbol']
        else:
            print(f"No ticker found for {company_name}")
            return None
    except Exception as e:
        print(f"Error: {e}")
        return None

# Ask user for company name
company_name = input("Enter the company name: ")

# Get ticker symbol for the company name
ticker_symbol = get_ticker_from_name(company_name)

if ticker_symbol:
    # Fetch the data for the given ticker symbol
    dat = yf.Ticker(ticker_symbol)
    
    # Fetch various data from yfinance
    print(f"Fetching data for: {ticker_symbol}")
    print("Company Info:")
    print(dat.info)
    
    print("\nCompany Calendar:")
    print(dat.calendar)
    
    print("\nAnalyst Price Targets:")
    print(dat.analyst_price_targets)
    
    print("\nQuarterly Income Statement:")
    print(dat.quarterly_income_stmt)
    
    print("\nHistorical Data (1 month):")
    print(dat.history(period='1mo'))
    
    print("\nOption Chain (first expiration):")
    print(dat.option_chain(dat.options[0]).calls)
    
    # Fetch data for multiple tickers
    tickers = yf.Tickers(ticker_symbol)
    print("\nMultiple Tickers Info:")
    print(tickers.tickers[ticker_symbol].info)
    
    # Download historical data for the company
    historical_data = yf.download([ticker_symbol], period='1mo')
    print("\nHistorical Data Downloaded:")
    print(historical_data)
else:
    print("Unable to fetch data.")
```
