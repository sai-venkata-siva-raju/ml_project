```python
import requests
from bs4 import BeautifulSoup
request = requests.get('https://www.moneycontrol.com/news/')
print(type(request))
print(request.text)
```

```python
soup = BeautifulSoup(request.text, 'html.parser')
print(type(soup))
print(soup.title)
```

```python
header3tags = soup.find_all('h3', attrs={'class': 'related_des'})
a = []
for header in header3tags:
    anchors = header.find_all('a')  # Call find_all on each individual h3 tag
    for anchor in anchors:
        a.append(anchor.text)
        print(a[-1])
```

```python
# lowercasetext = text.lower()
# translatedtext=lowercasetext.translate({ord(','): ' ', ord('!'): ' ', ord('@'): ' ',ord('.'): ' ',ord('/'): ' ',ord('\\'): ' ',ord(';'): ' ',ord('\''): ' ',ord('\"'): ' '})
# tokenizedtext = translatedtext.split()

# stop_words = ["i", "me", "my", "myself", "we", "our", "ours", "ourselves", "you", "your", "yours", "yourself",
#               "yourselves", "he", "him", "his", "himself", "she", "her", "hers", "herself", "it", "its", "itself",
#               "they", "them", "their", "theirs", "themselves", "what", "which", "who", "whom", "this", "that", "these",
#               "those", "am", "is", "are", "was", "were", "be", "been", "being", "have", "has", "had", "having", "do",
#               "does", "did", "doing", "a", "an", "the", "and", "but", "if", "or", "because", "as", "until", "while",
#               "of", "at", "by", "for", "with", "about", "against", "between", "into", "through", "during", "before",
#               "after", "above", "below", "to", "from", "up", "down", "in", "out", "on", "off", "over", "under", "again",
#               "further", "then", "once", "here", "there", "when", "where", "why", "how", "all", "any", "both", "each",
#               "few", "more", "most", "other", "some", "such", "no", "nor", "not", "only", "own", "same", "so", "than",
#               "too", "very", "s", "t", "can", "will", "just", "don", "should", "now"]
# final_words = [word for word in tokenizedtext if word not in stop_words]

# final_words

from textblob import TextBlob

def classify_sentiment(text):
    blob = TextBlob(text)
    
    polarity = blob.sentiment.polarity
    
    if polarity > 0:
        return 'positive'
    elif polarity < 0:
        return 'negative'
    else:
        return 'neutral'


headline = "Stocks rally as markets show strong growth"
print(f"Sentiment: {classify_sentiment(headline)}")

import nltk
nltk.download('vader_lexicon')
from vaderSentiment.vaderSentiment import SentimentIntensityAnalyzer

def classify_sentiment_vader(text):
    # Initialize VADER SentimentIntensityAnalyzer
    analyzer = SentimentIntensityAnalyzer()
    
    # Get sentiment scores
    sentiment = analyzer.polarity_scores(text)
    
    # Classify based on compound score
    compound_score = sentiment['compound']
    if compound_score > 0.05:
        return 'positive'
    elif compound_score < -0.05:
        return 'negative'
    else:
        return 'neutral'

# Example usage:
headlines = a

for headline in headlines:
    print(f"Headline: {headline} -> Sentiment: {classify_sentiment_vader(headline)}")
```

```python

import matplotlib.pyplot as plt
import seaborn as sns

headlines = [
    "Stock prices soar amid positive earnings reports",
    "Company X faces huge losses as competitors gain",
    "Investors uncertain about the future of the economy",
    "Tech sector booming with new innovations",
    "Global markets plunge due to political instability"
]

# Classify sentiments
sentiments = [classify_sentiment_vader(headline) for headline in headlines]

# Plot the sentiment distribution
sns.countplot(x=sentiments)
plt.title('Sentiment Analysis of News Headlines')
plt.xlabel('Sentiment')
plt.ylabel('Count')
plt.show()
```
